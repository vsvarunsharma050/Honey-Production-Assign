```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('honeyproduction 1998-2021.csv')
df.head()
```

```python
##1. How has honey production yield changed from 1998 to 2021?
Y_Prod = df.groupby('year')['yieldpercol'].mean().reset_index()
Y_Prod.sort_values('yieldpercol',ascending=False)
```

```python
##1. How has honey production yield changed from 1998 to 2021?
plt.figure(figsize=(10,4))
sns.pointplot(data=Y_Prod,x='year',y='yieldpercol',ci=None)
plt.title("Average Honey Production per colony over the Years(In Pounds)")
plt.xticks(rotation=90)
plt.grid('gridlines')
plt.show()
```

##In the above graph we can see that there has been a sharp decline in the production of Honey per colony after year "2000"  which has been constantly going down across the years and has reached to nearly the 60% of what it was at 1998.

```python
## Ques 2. Over time, what are the major production trends across the states?

## Solution: We will see it by comparing different sets of 5 states in each set for a better visualization & understanding.

df['State'].unique()
```

```python
a1= df[df['State'].isin(['Alabama','Montana','New Jersey', 'New York','Colorado'])]
a2= df[df['State'].isin(['South Dakota','Ohio','Texas', 'Missouri','Wisconsin'])]

```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data=a1,x='year',y='totalprod',hue='State')
plt.title('State wise')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
plt.figure(figsize=(10,4))
sns.lineplot(data=a2,x='year',y='totalprod',hue='State')
plt.title('State wise')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

# I have taken 10 states out of the whole data and made assumptions on these basis. 
# 7 out of 10 states are in clear declining graph. So, we can say that total production has gone down in every state signinficantly.

```python
## Ques.3: Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, 
# which was when concern over Colony Collapse Disorder spread nationwide?

db4_2006 = df[df['year']<2006]
db4_2006
```

```python
db4_2006['year'].value_counts()
```

```python
db1= db4_2006.groupby('year')['numcol','yieldpercol'].sum().reset_index()
db1
```

```python
## Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006,
#which was when concern over Colony Collapse Disorder spread nationwide?
sns.lineplot(data=db1,x='year',y='numcol') 
plt.title('No.of Colonies over the Years before 2006',size=15,color='red')
plt.ylabel('Number of Colonies',size=15,color='red')
plt.xlabel('Year',size=15,color='red')
plt.show()
sns.lineplot(data=db1,x='year',y='yieldpercol') 
plt.title('Average Honey Yield per colony before 2006',size=15,color='red')
plt.ylabel('Average Honey Yield in a Colony',size=15,color='red')
plt.xlabel('Year',size=15,color='red')
plt.show()
```

## We can clearly see from the above plotted graphs that the decline in number of colonies and yield per colony had already started before year 2006 when they started noticing the decline. From here, I can say that the real decline was triggered at year - 2000-2001 as per the given data and graph.<br> 

```python
## Ques.4: Are there any patterns that can be observed between total honey production and value of production every year?
b1= df.groupby('year')['prodvalue','totalprod'].sum().reset_index()
b1
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=b1,x='year',y='prodvalue',color='blue',label='prodvalue')
sns.barplot(data=b1,x='year',y='totalprod',color='red',label='totalprod',alpha=0.5)
plt.title('Total Production vs Production Value',color='red',size=15)
plt.ylabel('Production and Value of Production',color='red',size=15)
plt.xlabel('Year',color='red',size=15)
plt.xticks(rotation=90)
plt.grid('gridlines')
plt.legend(loc="best")
plt.show()
```

## We can see in this graph that after 2008 there has been a rise in gap between the Total Production and Value of prodution which has doubled in last 8 years as per the given data.

```python
## Ques.5: How has the value of production, which in some sense could be tied to demand, changed every year?
Pvalue= df.groupby('year')['prodvalue'].sum().reset_index()
Pvalue.sort_values('prodvalue',ascending=False)
```

```python
## Ques. How has the value of production, which in some sense could be tied to demand, changed every year?
plt.figure(figsize=(10,4))
sns.barplot(data=Pvalue,x='year',y='prodvalue')
plt.title('Value of Production Year Wise (in $)',color='red',size=15)
plt.xticks(rotation=90)
plt.xlabel('Years',color='red',size=15)
plt.ylabel('Production Value',color='red',size=15)
plt.show()
```

## Above Graph shows us that Production Value of the Honey increased after Year '2007' and has reached highest till 38 million USD in 2014  after that in recent years it is constant at around 29 to 32 million USD  a year.
